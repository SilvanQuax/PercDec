```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.optimize as sio

R=[]
R.append(np.genfromtxt('result/supervised/mnist_sup_dn3_h5_t10_class_accuracy', delimiter=','))
R.append(np.genfromtxt('result/supervised/mnist_sup_cn2_h5_t10_class_accuracy', delimiter=','))
R.append(np.genfromtxt('result/supervised/mnist_sup_dn2_h10_t10_class_accuracy', delimiter=','))
R.append(np.genfromtxt('result/supervised/mnist_sup_cn2_h10_t10_class_accuracy', delimiter=','))
R.append(np.genfromtxt('result/supervised/mnist_sup_dn3_h30_t10_class_accuracy', delimiter=','))
R.append(np.genfromtxt('result/supervised/mnist_sup_cn3_h30_t10_class_accuracy', delimiter=','))

noise=100-np.logspace(np.log10(100),np.log10(3.125),num=6)

fitvar=[[] for i in range(6)]
h_all=[5,10,30]
x=np.arange(10)
x1=np.arange(0,10,0.01)

def f1(x,b1,b2,b3):
    return -(b3-b1)*np.exp(b2*x)+b3

plt.figure(figsize=(7,8))
hidden=[5,5,10,10,30,30]

for iii in xrange(6):
    plt.subplot(321+iii)
    for ii in xrange(6):
        y=R[iii][ii,:].T
        fitvar[iii].append(sio.curve_fit(f1,x,y,p0=[0.1,-1,1],ftol=1e-06,xtol=1e-06))
        plt.plot(x1,f1(x1,*fitvar[iii][ii][0]))
    plt.gca().set_prop_cycle(None)
    for ii in xrange(6):
        y=R[iii][ii,:].T
        plt.plot(x,y,'o')
    plt.title('# hidden = '+ str(hidden[iii]))
    plt.xlabel('Time steps')
    plt.ylabel('Accuracy')
plt.tight_layout()

plt.show()
```

```python
trial_length = np.zeros((6,6))
x2=np.arange(0,100,0.1)

for iii in xrange(6):
    for ii in xrange(6):
        A=f1(x2,*fitvar[iii][ii][0])
        trial_length[iii,ii]=np.nonzero(A>0.95*np.max(A))[0][0]/10.0
    

plt.figure(figsize=(7,2.3))
plt.clf()
plt.subplot(121)
for ii in xrange(6):
    plt.plot(x1,f1(x1,*fitvar[0][ii][0]),linewidth=2.0)

plt.gca().set_prop_cycle(None)
for ii in xrange(6):
    y=R[0][ii,:].T
    plt.plot(x,y,ls='None',marker='o')

    

plt.xlabel('Time steps')
plt.ylabel('Accuracy')
aa=str(noise)[1:-1].split()[:]
aa=['N = ' + str(i) + '%' for i in aa]
plt.legend(aa,bbox_to_anchor=(1.14,1), loc="upper left")

plt.subplot(122)
plt.plot(noise,trial_length[0:5:2].T,marker='o',linewidth=2.0)

plt.xlabel('Noise')
plt.ylabel('Time steps')

plt.tight_layout()

plt.subplots_adjust(wspace=1.5, hspace=None)

plt.show()

plt.figure(figsize=(8,2.8))

for ii in xrange(3):
    plt.subplot(131+ii)
    plt.plot(noise,trial_length[ii*2].T,marker='o',linewidth=2.0)
    plt.xlabel('Noise')
    plt.ylabel('Time steps')
    plt.title('# hidden = '+ str(hidden[ii*2]))
    plt.ylim(0,20)

plt.tight_layout()

plt.show()

###

plt.figure(figsize=(8,2.8))
for ii in xrange(3):
    p=np.polyfit(np.log(100-noise),trial_length[ii*2],1)
    x_fit=np.log(100-noise)
    y_fit=x_fit*p[0]+p[1]
    
    plt.subplot(131+ii)

    plt.semilogx(100-noise,trial_length[ii*2].T,'ko')
    plt.semilogx(100-noise,y_fit,'k')

    plt.xlim(1,100)
    plt.xlabel('Signal strength (100 - % noise)')
    plt.ylabel('Time steps')
    plt.title('# hidden = '+ str(hidden[ii*2]))
    plt.ylim(0,20)
plt.tight_layout()

plt.show()
```
